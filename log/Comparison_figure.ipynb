```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df_baseline = pd.read_csv('training_log_baseline.csv')
df_2frame = pd.read_csv('training_log_2frame.csv')
df_3frame = pd.read_csv('training_log_3frame.csv')
df_5frame = pd.read_csv('training_log_5frame.csv')
df_LRCN = pd.read_csv('training_log_LRCN.csv')
df_multiTRN = pd.read_csv('training_log_multiscaleTRN.csv')
```

```python
num_epoch = df_baseline['epoch']
```

```python
val_baseline = df_baseline['val_acc']
val_2frame = df_2frame['val_acc']
val_3frame = df_3frame['val_acc']
val_5frame = df_5frame['val_acc']
val_LRCN = df_LRCN['val_acc']
val_multiTRN = df_multiTRN['val_acc'][0:30]
```

```python
plt.figure(figsize=(18,6))
plt.plot(num_epoch, val_baseline)
plt.plot(num_epoch, val_2frame)
plt.plot(num_epoch, val_3frame)
plt.plot(num_epoch, val_5frame)
plt.plot(num_epoch, val_LRCN)
plt.plot(num_epoch, val_multiTRN)
plt.legend(["Baseline","2-Frame TRN","3-Frame TRN","5-Frame TRN","LRCNs","MultiScale TRN"],loc='lower right')
plt.xlabel('Epoch')
plt.ylabel('Valiadation Accuracy')
plt.title('Valiadation accuracy vs number of epochs')
```
